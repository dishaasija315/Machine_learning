```python
list=[[2,3,4],[1,24,16,64],[2,6,9,12]]
sortlist = lambda x : [sorted(i) for i in x ]
print (sortlist(list))

secondlargest=lambda x : [y[-2] for y in sortlist(x) ]
print (secondlargest(list))
```
